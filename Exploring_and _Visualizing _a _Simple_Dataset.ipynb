```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the iris dataset from seaborn
iris = sns.load_dataset('iris')
```

```python
# Basic information
print("Shape of dataset:", iris.shape)
print("\nColumn names:", iris.columns.tolist())
print("\nFirst 5 rows:")
print(iris.head())
```

```python
# Info about dataset
print("\nDataset Info:")
print(iris.info())
```

```python
# Summary statistics
print("\nSummary Statistics:")
print(iris.describe())
```

```python
# Scatter plot: Sepal Length vs Petal Length
plt.figure(figsize=(8, 5))
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species', palette='Set1')
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
# Histograms of numerical features
iris.hist(figsize=(10, 8), color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Iris Features', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Box plots for outlier detection
plt.figure(figsize=(12, 6))
sns.boxplot(data=iris, orient='h', palette='Pastel1')
plt.title('Boxplots of Iris Features')
plt.show()
```
